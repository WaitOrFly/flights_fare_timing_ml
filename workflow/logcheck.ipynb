```python
import boto3
from datetime import datetime, timedelta

def check_endpoint_logs(endpoint_name):
    logs_client = boto3.client('logs', region_name='us-east-1')  #  
    log_group = f'/aws/sagemaker/Endpoints/{endpoint_name}'
    
    try:
        #   
        streams = logs_client.describe_log_streams(
            logGroupName=log_group,
            orderBy='LastEventTime',
            descending=True,
            limit=3
        )
        
        print(f"=== Endpoint: {endpoint_name} ===\n")
        
        for stream in streams['logStreams']:
            print(f"\n--- {stream['logStreamName']} ---")
            
            #   
            events = logs_client.get_log_events(
                logGroupName=log_group,
                logStreamName=stream['logStreamName'],
                startFromHead=False,
                limit=50
            )
            
            for event in events['events']:
                print(event['message'])
                
    except Exception as e:
        print(f"   : {e}")
        print("AWS Console  :")
        print(f"CloudWatch  Logs  {log_group}")

# 
check_endpoint_logs('flight-price-endpoint-1768021785-74f2')
```

```python
import boto3

def check_latest_endpoint_logs():
    sm_client = boto3.client('sagemaker')
    logs_client = boto3.client('logs')
    
    endpoints = sm_client.list_endpoints(
        SortBy='CreationTime',
        SortOrder='Descending',
        MaxResults=1
    )
    
    if not endpoints['Endpoints']:
        print("   .")
        return
    
    latest_endpoint = endpoints['Endpoints'][0]['EndpointName']
    print(f"=== : {latest_endpoint} ===")
    print(f"Status: {endpoints['Endpoints'][0]['EndpointStatus']}\n")
    
    log_group = f'/aws/sagemaker/Endpoints/{latest_endpoint}'
    
    try:
        streams = logs_client.describe_log_streams(
            logGroupName=log_group,
            orderBy='LastEventTime',
            descending=True,
            limit=10
        )
        
        for stream in streams['logStreams']:
            stream_name = stream['logStreamName']
            
            # container_1 container_2 
            if 'container' in stream_name:
                print(f"\n--- {stream_name} ---")
                
                events = logs_client.get_log_events(
                    logGroupName=log_group,
                    logStreamName=stream_name,
                    limit=100
                )
                
                #    
                for event in events['events']:
                    msg = event['message']
                    if any(keyword in msg.lower() for keyword in ['error', 'exception', 'traceback', 'failed', 'cannot']):
                        print(msg)
                    
    except Exception as e:
        print(f"  : {e}")

check_latest_endpoint_logs()
```
